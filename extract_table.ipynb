```python
from bs4 import BeautifulSoup
import pandas as pd

with open('ER_data.html', 'r') as file:
    html = file.read()

# Parse the page content using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Find the table with the class name 'wiki_table sortable searchable'
table = soup.find('table', {'class': 'wiki_table sortable searchable'})

# Extract table headers
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())

# Extract table rows and data
rows = []
for tr in table.find_all('tr')[1:]:  # Skip the first row (headers)
    cells = tr.find_all('td')
    row_data = [cell.get_text(strip=True) for cell in cells]
    rows.append(row_data)

# Create a pandas DataFrame
df = pd.DataFrame(rows, columns=headers)

# Display the DataFrame
df.info()
```

```python
# Python code to save DataFrame as JSON
df.to_json("ER_cookbooks.json", orient='records')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

with open('ER_data.html', 'r') as file:
    html = file.read()

# Parse the page content using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Find the table with the class name 'wiki_table sortable searchable'
table = soup.find('table', {'class': 'wiki_table sortable searchable'})

# Extract table headers
headers = ['Name', 'Crafting Repertoire', 'Location', 'Map Link', 'Latitude', 'Longitude', 'ID']
rows = []

# Loop through the table rows and extract data
for tr in table.find_all('tr')[1:]:  # Skip the first row (headers)
    cells = tr.find_all('td')
    
    # Extract Name, Crafting Repertoire, and Location
    name = cells[0].text.strip()
    crafting_repertoire = cells[1].text.strip()
    location = cells[2].text.strip()

    # Extract the href attribute
    map_link_tag = cells[2].find('a', href=True)
    map_link = map_link_tag['href'] if map_link_tag else None

    # Extract the id, lat, and lng using regex
    if map_link:
        print(map_link)
        match = re.search(r'id=(\d+)&lat=([-+]?\d*\.\d+|\d+)&lng=([-+]?\d*\.\d+|\d+)', map_link)
        if match:
            map_id = match.group(1)
            lat = match.group(2)
            lng = match.group(3)
        else:
            map_id = lat = lng = None
    else:
        map_id = lat = lng = None

    # Append data to the rows list
    rows.append([name, crafting_repertoire, location, map_link, lat, lng, map_id])

# Create a pandas DataFrame
df = pd.DataFrame(rows, columns=headers)

df
```
